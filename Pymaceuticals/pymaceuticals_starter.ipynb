## Observations and Insights 

Observations

Observation 1:  From the statistics table, Capomulin shows on average a smaller average tumor volume than all other Drug 
                Regimens except Ramicane.
Observation 2:  From the "Number of Mice per Timepoint" bar chart, it appears Capomulin has the most significant impact 
                on reducing mice deaths. This is evident where the 45 Timepoint Mouse Count is higher than the other 
                Drug Regimen at the 45 Timepoint.
Observation 3:  Based on the "Mouse Weight versus Average Tumor Volume for the Capomulin regimen" scatter plot, it can 
                be inferred that there is a strong correlation between Mouse Weight and the Average Tumor Volume. 
                The heavier the mouse the larger the Tumor Volume.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
Capomulin_study_data = pd.merge(mouse_metadata, study_results, how="outer", on="Mouse ID")
Capomulin_study_data
```

```python
# Checking the number of mice in the DataFrame.
mice_group = Capomulin_study_data.groupby("Mouse ID")
number_of_mice = len(mice_group["Mouse ID"])
number_of_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mouse = Capomulin_study_data[Capomulin_study_data.duplicated(keep="first")] 
print(duplicate_mouse["Mouse ID"], duplicate_mouse["Timepoint"])
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse = Capomulin_study_data[Capomulin_study_data.duplicated()] 
duplicate_mouse
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume 
#for each regimen

# Group by regimen
drug_regimen_group = Capomulin_study_data_clean.groupby(["Drug Regimen"])

#Calculate mean for of the tumor volume per regimen

drug_regimen_mean = round(drug_regimen_group["Tumor Volume (mm3)"].mean(), 2)
drug_regimen_mean

#Calculate the median of the tumor volume per regimen

drug_regimen_median = round(drug_regimen_group["Tumor Volume (mm3)"].median(), 2)
drug_regimen_median

#Calculate the variance of the tumor volume per regimen

drug_regimen_variance = round(drug_regimen_group["Tumor Volume (mm3)"].var(), 2)
drug_regimen_variance

#Calculate the standard deviation of the tumor volume per regimen

drug_regimen_std_dev = round(drug_regimen_group["Tumor Volume (mm3)"].std(), 2)
drug_regimen_std_dev

#Calculate the SEM of the tumor volume per regimen
drug_regimen_sem = round(drug_regimen_group["Tumor Volume (mm3)"].sem(), 2)
drug_regimen_sem

# Generate summary table of the statistics
Tumor_Volume_summary_table_groupby = pd.DataFrame({"Tumor Volume (mm3) Mean": drug_regimen_mean, 
                                         "Tumor Volume (mm3) Median": drug_regimen_median, 
                                         "Tumor Volume (mm3) Variance": drug_regimen_variance, 
                                         "Tumor Volume (mm3) Standard Deviation": drug_regimen_std_dev, 
                                         "Tumor Volume (mm3) Standard Error of the Mean": drug_regimen_sem})

#Show table
Tumor_Volume_summary_table_groupby
```

## Correlation and Regression

```python
#Calculate the correlation coefficient and linear regression model for
#mouse weight and average tumor volume for the Capomulin regimen

#Collect weight and volume data for correlation coefficient calulation
weight = scatter_df["Weight (g)"] 
volume = scatter_df["Average Tumor Volume"]

#Calculating correlation coefficient
correlation = st.pearsonr(weight,volume)
print(f"The correlation between both factors is {round(correlation[0],2)}")

```

```python
#Calculate linear regression model

#Generate tumor_volume and mouse_weight
tumor_volume = scatter_df["Average Tumor Volume"]
mouse_weight = scatter_df["Weight (g)"]


(slope, intercept, rvalue, pvalue, stderr) = st.linregress(tumor_volume, mouse_weight)

#Generate line equation
regress_values = tumor_volume * slope + intercept

#Create line equation for annotating to plot
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

#Scatter plot via py.plot
plt.scatter(tumor_volume, mouse_weight, marker="o", facecolors="orange", edgecolors="black")
plt.plot(tumor_volume,regress_values,"b-")
plt.title("Mouse Weight versus Average Tumor Volume for the Capomulin regimen")
plt.xlabel('Average Tumor Volume')
plt.ylabel('Mouse Weight (g)')

#Annotate line equation
plt.annotate(line_eq,(40,20),fontsize=15,color="red")

# Print out the r-squared value along with the plot.
print(f"The r-squared is: {rvalue**2}")

plt.show()

```
